```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load sample image
img = cv2.imread("../data/training/image_2/um_000000.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Load ground truth mask
mask = cv2.imread("../data/training/gt_image_2/um_road_000000.png")
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)

# Create binary road mask
road = mask[:, :, 2] > 0

# Create overlay
overlay = img.copy()
overlay[road] = [255, 0, 0]

# Show image, mask, and overlay
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title("KITTI Sample Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(road, cmap="gray")
plt.title("Road Mask")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(overlay)
plt.title("Overlay")
plt.axis("off")

plt.tight_layout()
plt.show()

# Compute road pixel ratio
road_pixels = road.sum()
total_pixels = road.size
ratio = road_pixels / total_pixels

print(f"Road pixel ratio: {ratio:.4f}")
print(f"Road area percentage: {ratio * 100:.2f}%")
```
